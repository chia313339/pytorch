# 傅立葉轉換

```python
import matplotlib.pyplot as plt 
from scipy.fftpack import fft,ifft
```

```python
def fft_process(data):
    data = np.array(data)
    data_fft = fft(data)
    data_fft = abs(data_fft)/len(data_fft)
    return data_fft[0:int((data.shape[0])/2)]
```

```python
ecg_train_eachfft = pd.DataFrame()
```

```python
for index,uid in enumerate(train_metadata['UID']):
    print('處理 UID:',uid,',進度',index+1,'/',train_metadata['UID'].shape[0])
    ecg_tmp = pd.read_csv('ecg/{}.csv'.format(uid))
    ecg_fft = ecg_tmp.apply(lambda x: fft_process(x))
    ecg_vec = ecg_fft.values.reshape(1,6000)
    ecg_train_eachfft = pd.concat([ecg_train_eachfft, pd.DataFrame(ecg_vec)], ignore_index=True, sort=False)
```

```python
ecg_train_eachfft.to_csv('ecg_train_eachfft.gz', compression='gzip', index=False)
```

```python
ecg_test_eachfft = pd.DataFrame()
```

```python
for index,uid in enumerate(test_metadata['UID']):
    print('處理 UID:',uid,',進度',index+1,'/',test_metadata['UID'].shape[0])
    ecg_tmp = pd.read_csv('ecg/{}.csv'.format(uid))
    ecg_fft = ecg_tmp.apply(lambda x: fft_process(x))
    ecg_vec = ecg_fft.values.reshape(1,6000)
    ecg_test_eachfft = pd.concat([ecg_test_eachfft, pd.DataFrame(ecg_vec)], ignore_index=True, sort=False)
```

```python
ecg_test_eachfft.to_csv('ecg_test_eachfft.gz', compression='gzip', index=False)
```
